```python
import cv2
```

```python
import numpy as np
```

```python
import matplotlib.pyplot as plt
```

```python
img=np.zeros((513,513,3), np.uint8)
```

```python
cv2.rectangle(img,(0,0),(513,171),(255,165,0),-1) #orange 
```

```python
cv2.rectangle(img,(0,171),(512,342),(255, 255, 255),-1) #white
```

```python
cv2.circle(img,(256,256), 80, (0,0,255), -1)
```

```python
cv2.rectangle(img,(0,342),(512,512),(0,255,0),-1) #green
```

```python
cv2.imshow('FLAG',img)
cv2.waitKey()
cv2.destroyAllWindows() 
```

```python
dog=cv2.imread('golden.jpeg')
elephant=cv2.imread('elephant.jpeg')
```

```python
dog.shape
```

```python
elephant.shape
```

```python
dog_crop=dog[0:225,0:192] 
elephant_crop=elephant[0:225,0:192]
Image_combine=cv2.hconcat([dog_crop,elephant_crop])
```

```python
plt.imshow(dog)
```

```python
plt.imshow(elephant)
```

```python

cv2.imshow("Combined_image",Image_combine)
cv2.waitKey()
cv2.destroyAllWindows()
```
